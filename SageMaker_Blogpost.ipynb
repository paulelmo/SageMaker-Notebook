**2.	Options when creating a new notebook**

**2.1.	Notebook Instance Type**
Sagemaker provides a wide selection of instance types optimized to fit different machine learning (ML) use cases. The differ in the way the work is done. By using a mix of CPU, GPU, memory, and networking capacity. It’s up to you to choose the right instance type. It’s even possible to change the instance type on the fly. If your project grows bigger and bigger you can scale the resources bigger and bigger.

![](images/InstanceExample.png "Instance Types")

**Example:** 
> ml.t2.medium = machine learning standard second generation medium size

> ml.r5.2xlarge = machine learning memory optimized fifth generation extra large

**2.2.	Elastic Inference**

To speed up your learning progress you could choose an GPU-powered Instance type. On the other hand, GPU-powered instances are more expensive than CPU- or RAM-powered instances. But these are often too high in latency for real-time actions. If you don’t want a compromise you should look at Elastic Inference. Elastic Inference lets you attach just the right amount of GPU-powered inference acceleration to a Sagemaker Model. Pick the best CPU instance type for your application, attach the right amount of GPU acceleration and get the best of both worlds.

**2.3.	Permissions and Encryption** 

IAM stands for Identity and Access Management. IAM is a service that helps you securely control access to AWS resources. You use IAM to control who is authenticated (signed in) and authorized (has permissions) to use resources.
When registered for your first account you are starting with an all-access Single-Sign-In Account (also called root user). AWS strongly recommends not to use the root user for everyday tasks, not even administrative ones.  Instead you should start with individual IAM-roles. You can also encrypt your notebook. For further Details have a look at the aws Documentation.  

![](images/PermissionsEncryption.png "Permissions and Encryption")

**2.4.	Optional Customization**

Amazon Virtual Private Cloud (Amazon VPC) enables you to launch AWS resources into a virtual network that you've defined. This virtual network closely resembles a traditional network that you'd operate in your own data center, with the benefits of using the scalable infrastructure of AWS. 

![](images/OptionalCustomization.png "Optional Customization")

Associate Git repositories with your notebook instance to save your notebooks in a source control environment that persists even if you stop or delete your notebook instance. You can associate one default repository and up to three additional repositories with a notebook instance. The repositories can be hosted in AWS CodeCommit, GitHub or on any other Git server. It can helpful for persistence, collaboration or learning. 

**2.5.	Finished!**

If you are done with setting the parameters, you can click on *create notebook instance*. It may take a while if your notebook is active. Usually some minutes. Take a break, have yourself a Cup of Coffee and get ready for the next step.
